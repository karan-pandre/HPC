```python
import numpy as np
import pandas as pd
import time

def check_mkl():
    if np.__config__.get_info('blas_mkl_info'):
        print("NumPy is using Intel MKL.")
    else:
        print("NumPy is not using Intel MKL.")

def compute_rolling_statistics(data, column, window_size):
    start_time = time.time()
    rolling_mean = data[column].rolling(window=window_size).mean()
    rolling_std = data[column].rolling(window=window_size).std()
    end_time = time.time()
    print(f"Time taken to compute rolling statistics: {end_time - start_time:.4f} seconds")
    return rolling_mean, rolling_std

# Configuration Check
check_mkl()

# Load the dataset
data = pd.read_csv('/Users/karanpandre/Documents/INTERNSHIP/HPC /case_study(28/6)/bank_customer.csv')

# Print column names
print("Column names in the dataset:")
print(data.columns)

# Select a numeric column (replace 'BloodSugar' with an actual numeric column name)
numeric_column = 'customer_id'

# Check if the column exists
if numeric_column in data.columns:
    # Compute Rolling Statistics
    window_size = 50
    rolling_mean, rolling_std = compute_rolling_statistics(data, numeric_column, window_size)

    # Display Results
    print(f"\nOriginal Data - {numeric_column} (first 10 rows):")
    print(data[numeric_column].head(10))

    print(f"\nRolling Mean of {numeric_column} (first 10 rows):")
    print(rolling_mean.head(10))

    print(f"\nRolling Standard Deviation of {numeric_column} (first 10 rows):")
    print(rolling_std.head(10))
else:
    print(f"Column '{numeric_column}' not found in the dataset.")
```
